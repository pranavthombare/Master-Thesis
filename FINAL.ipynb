```python
face_model = load_model("facenet_keras.h5")
```

```python
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
#FUNCTION TO ENCODE THE IMAGE
def img_encoding(image):
    img=np.resize(image,(160,160,3))
    img=img.reshape(1,160,160,3)
    #img_enc= (face_model.output,feed_dict={face_model.input:img})
    img_enc = face_model.predict(img)
    return img_enc
```

```python
#FUNCTION TO VERIFY THE PERSON IDENTITY
def verify(check_img,identity,database):
    true_img=database[identity]
    err=np.linalg.norm(true_img.T - check_img.T)
    if (err<1.6):
        print('welcome '+str(identity))
        print(err)
    else:
        print('you are not allowed,please leave')
        print(err)
```

```python
#DATABASE
def img_database(image,name):
    database[name]=img_encoding(image)
```

```python
#MAIN FUNCTION 
def main_fun(input_img,name):
    check_img1=img_encoding(input_img)    
    verify(check_img1,name,database)
```
